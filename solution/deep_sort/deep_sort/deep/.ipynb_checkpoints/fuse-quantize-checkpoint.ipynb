```python
def generate_activation_ballperson():
    from glob import glob
    import os
    import numpy as np
    import cv2
    import random
    
    root_dir=r'C:\Workspace\study-projects\LPCV\ballPerson\images\train'
    all_files = list(glob(os.path.join(root_dir, '*')))
    random.shuffle(all_files)
    
    top_n = 4
    top_n_files = all_files[:top_n]
    
    size = (640, 640)
    
    image_batch = []
    for file in top_n_files:
        image = cv2.imread(file)
        image = cv2.resize(image, size)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = (image / 255.0).astype(np.float32)
        image = torch.from_numpy(image).permute(2, 0, 1).unsqueeze(0)
        image_batch.append(image)

    image_batch = torch.cat(image_batch, dim=0)
    
    return image_batch

sample_activation_dataset = generate_activation_ballperson()
print(sample_activation_dataset.shape)

# model = model.quantize(sample_activation_dataset)
```

```python
from yolox.data.data_augment import ValTransform
preproc = ValTransform(legacy=False)

def postprocess(prediction, num_classes, conf_thre=0.7, nms_thre=0.45):
    box_corner = prediction.new(prediction.shape)
    box_corner[:, :, 0] = prediction[:, :, 0] - prediction[:, :, 2] / 2
    box_corner[:, :, 1] = prediction[:, :, 1] - prediction[:, :, 3] / 2
    box_corner[:, :, 2] = prediction[:, :, 0] + prediction[:, :, 2] / 2
    box_corner[:, :, 3] = prediction[:, :, 1] + prediction[:, :, 3] / 2
    prediction[:, :, :4] = box_corner[:, :, :4]

    output = [None for _ in range(len(prediction))]
    
    print('prediction:', prediction)
    for i, image_pred in enumerate(prediction):

        # If none are remaining => process next image
        if not image_pred.size(0):
            continue
        # Get score and class with highest confidence
        class_conf, class_pred = torch.max(image_pred[:, 5: 5 + num_classes], 1, keepdim=True)
        
        confidence_map = image_pred[:, 4] * class_conf.squeeze()
        conf_mask = (confidence_map >= conf_thre).squeeze()
        
        print("Max confidence:", torch.max(confidence_map), ", Min confidence:", torch.min(confidence_map))
        
        # print("Confidence mask (Object count):", conf_mask)
        # Detections ordered as (x1, y1, x2, y2, obj_conf, class_conf, class_pred)
        detections = torch.cat((image_pred[:, :5], class_conf, class_pred.float()), 1)
        detections = detections[conf_mask]
        if not detections.size(0):
            print("Zero", end=' ')
            continue

        nms_out_index = torchvision.ops.batched_nms(
            detections[:, :4],
            detections[:, 4] * detections[:, 5],
            detections[:, 6],
            nms_thre,
        )
        detections = detections[nms_out_index]
        if output[i] is None:
            print("Output is None, detections=", detections)
            output[i] = detections
        else:
            print("Output=", output[i], ", detections=", detections)
            output[i] = torch.cat((output[i], detections))

    return output

def inference_batched(imgs, model):
    img_infos = []

    imgs_batched = []
    for img in imgs:
        img_info = {"id": 0}
        img_info["file_name"] = None

        height, width = img.shape[:2]
        img_info["height"] = height
        img_info["width"] = width

        ratio = min(640 / img.shape[0], 640 / img.shape[1])
        img_info["ratio"] = ratio

        img, _ = preproc(img, None, (640, 640))
        img = torch.from_numpy(img).unsqueeze(0)

        imgs_batched.append(img)
        img_infos.append(img_info)

    imgs_batched = torch.cat(imgs_batched, 0)  # batch-dim concat

    with torch.no_grad():
        outputs_batched = model(imgs_batched)
        
    return outputs_batched, img_infos
```

```python
from yolox.predictor import Predictor
from yolox.data.datasets.lpcvloader import LoadImages

BALLPERSON_CLASSES = (
    "person",
    "sports ball"
)

dataset = LoadImages('../../videos/4p1b_01A2/4p1b_01A2.m4v', img_size=640, batch_size=4, skip_frames=0, stride=32, multithreaded=False)

for i, (path, im0sb) in enumerate(dataset):
    if i >= 5:
        break
    preds, img_infos = inference_batched(im0sb, model)
    preds = postprocess(
        preds, 2, 0.35, 0.35
    )
    print('len(preds):', len(preds), preds[0])

# del dataset, predictor
```
